rice production analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib as mt
import matplotlib.pyplot as plt

data = pd.read_csv('Sugarcane_Production.csv')
data.head(10)
```


----------------------------------------------------------data cleaning-------------------------------------------------------------

```python
data['Production (Tons)'] = data['Production (Tons)'].str.replace(".","")
data['Production per Person (Kg)'] = data['Production per Person (Kg)'].str.replace(",","")
data['Acreage (Hectare)'] = data['Acreage (Hectare)'].str.replace(".","")
data['Yield (Kg / Hectare)'] = data['Yield (Kg / Hectare)'].str.replace(",","")
data.head(10)
```

renaming column

```python
data.rename(columns={"Production (Tons)" : "Production(Tons)"} , inplace=True)
data.rename(columns={"Production per Person (Kg)" : "ProductionperPerson(Kg)"} , inplace=True)
data.rename(columns={"	Acreage (Hectare)" : "	Acreage(Hectare)"} , inplace=True)
data.rename(columns={"Yield (Kg / Hectare)" : "Yield(Kg/Hectare)"} , inplace=True)
data.head(10)
```

checking if any null data 

```python
data.isna().sum()
data[data['Acreage (Hectare)'].isnull()]
```

deleting mislenious columns

```python
data.drop(["Unnamed: 0"], axis=1, inplace=True )
data.head(10)
```

checking data types of columns and convert them into float type

```python
data.dtypes
data['Production(Tons)'] = data['Production(Tons)'].astype(float)
data['ProductionperPerson(Kg)'] = data['ProductionperPerson(Kg)'].astype(float)
data['Acreage (Hectare)'] = data['Acreage (Hectare)'].astype(float)
data['Yield(Kg/Hectare)'] = data['Yield(Kg/Hectare)'].astype(float)
data.dtypes
```


--------------------------------------------------------bivariate analysis-------------------------------------------------------------------

which country produces maximum sugercane

```python
data_new = data[['Country','Production(Tons)']].set_index('Country')
data_new
```

```python
data_new['Production(Tons)_persent'] = data_new['Production(Tons)'] * 100 / data_new['Production(Tons)'].sum()
data_new
```

make pie cahrt of top 10 country out of 99

```python
data_new['Production(Tons)_persent'].head(10).plot(kind='pie' ,autopct = '%.2f')
```

making baar chart of top 10 out of 99

```python
data_new['Production(Tons)'].head(10).plot(kind='bar')
```

bar chart using seaborn library

```python
ax = sns.barplot(data=data.head(10) , x = 'Country' , y='Production(Tons)')
```

```python
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

which country has highest land 

```python
data_acr = data.sort_values('Acreage (Hectare)',ascending=False)
ax = sns.barplot(data=data_acr.head(10) , x='Country' , y='Acreage (Hectare)')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

which country has highest yeild per heactare

```python
data_acr = data.sort_values('Yield(Kg/Hectare)',ascending=False)
ax = sns.barplot(data=data_acr.head(10) , x='Country' , y='Yield(Kg/Hectare)')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
plt.show()
```

using scatterplot

continent analysis

```python
data_continent = data.groupby('Continent').sum()
data_continent
```

which continent produces highest sugercane

```python
data_continent['Production(Tons)'].sort_values(ascending=False).plot(kind='bar')
```

do number of countries in a countinent effect production of sugercane

```python
data_continent['number_of_country'] = data.groupby('Continent').count()['Country']
data_continent
```

```python
cotinent_names = data_continent.index.to_list()
sns.lineplot(data=data_continent , x='number_of_country' , y = 'Production(Tons)')
plt.xticks(data_continent['number_of_country'],cotinent_names,rotation=90)
plt.show()
```
